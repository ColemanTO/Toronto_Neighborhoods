```python
file_name='https://raw.githubusercontent.com/ColemanTO/Toronto_Neighborhoods/b788a5a617dd72aa157ad56f032214ac83a61058/Toronto_Postal_Codes.csv'
df2=pd.read_csv(file_name)
```

Remove Unassigned Postal Codes from df2

```python
# simply drop whole row with Not Assigned in "Borough" column
df2 = df2[df2.Borough != 'Not assigned']

# reset index, because we dropped some rows
df2.reset_index(drop=True, inplace=True)
df2
```

Join the remaining Postal Codes with their latitude and longitude data

```python
df3 = pd.merge(df2, df1, on="Postal Code")
```

```python
df3
```

Show Neighbourhoods

```python
toronto_map = folium.Map(location=[43.65, -79.36], zoom_start=12)

# instantiate a feature group for the neighbourhoods in the dataframe
neighbourhoods = folium.map.FeatureGroup()

# loop through the 103 neighbourhoods and add each to the neighbourhoods feature group
for lat, lng, label, in zip(df3.Latitude, df3.Longitude, df3.Borough):
    neighbourhoods.add_child(
        folium.features.Marker(
            [lat, lng],
            popup=label,
        )
    )
    
# add neighbourhoods to map
toronto_map.add_child(neighbourhoods)     
    
toronto_map
```
